```python
import cv2 
import os, glob

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

image_dir = "images/hand_gestures_digits_0_5"


images = []

for file in glob.glob(os.path.join(image_dir, "*")):
    im = cv2.imread(file)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

    images.append(im)

```

### Plot all the base images in a subplot of 2 rows and 3 columns

```python
f, ((im0, im1, im2),(im3,im4,im5)) = plt.subplots(2, 3, figsize=(10,10))

im0.set_title('1')
im0.imshow(images[0], cmap='gray')
im1.set_title('2')
im1.imshow(images[1], cmap='gray')
im2.set_title('3')
im2.imshow(images[2], cmap='gray')

im3.set_title('4')
im3.imshow(images[3], cmap='gray')
im4.set_title('5')
im4.imshow(images[4], cmap='gray')
im5.set_title('6')
im5.imshow(images[5], cmap='gray')


```

```python

images_processed = []
kernel = np.ones((5,5), np.uint8)
lower_hue = np.array([0,0,0]) 
upper_hue = np.array([255,150,255])
for i in images:  
    hsv_im = cv2.cvtColor(i, cv2.COLOR_RGB2HSV)

    mask_hsv = cv2.inRange(hsv_im, lower_hue, upper_hue)
    mono = np.copy(i)
    mono[mask_hsv==0] = [0,0,0]

    mono0 = cv2.cvtColor(mono, cv2.COLOR_RGB2GRAY)
    mono1 = cv2.resize(mono0, (375, 500))
    mono2 = cv2.Canny(mono1, 50, 100)
    mono3 = cv2.dilate(mono2, kernel, iterations = 1)

    images_processed.append(mono3)


f, ((im0, im1, im2),(im3,im4,im5)) = plt.subplots(2, 3, figsize=(10,10))

im0.set_title('1')
im0.imshow(images_processed[0], cmap='gray')
im1.set_title('2')
im1.imshow(images_processed[1], cmap='gray')
im2.set_title('3')
im2.imshow(images_processed[2], cmap='gray')

im3.set_title('4')
im3.imshow(images_processed[3], cmap='gray')
im4.set_title('5')
im4.imshow(images_processed[4], cmap='gray')
im5.set_title('6')
im5.imshow(images_processed[5], cmap='gray')
    
```
